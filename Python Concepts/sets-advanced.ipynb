```python
# Create an empty set
s = set()

# Add elements to the set
s.add(1)
s.add(2)
print(s)  # Output: {1, 2}

# Clear all elements from the set
s.clear()
print(s)  # Output: set()

# Create a set and make a copy
s = {1, 2, 3}
sc = s.copy()
print(sc)  # Output: {1, 2, 3}

# Modify the original set and observe the copy
s.add(4)
print(s)  # Output: {1, 2, 3, 4}
print(sc)  # Output: {1, 2, 3}

# Find the difference between two sets
s1 = {1, 2, 3}
s2 = {1, 4, 5}
diff_set = s1.difference(s2)
print(diff_set)  # Output: {2, 3}

# Use difference_update to update a set with the difference
s1.difference_update(s2)
print(s1)  # Output: {2, 3}

# Discard an element from a set
s = {1, 2, 3, 4}
s.discard(2)
print(s)  # Output: {1, 3, 4}

# Find the intersection of two sets
s1 = {1, 2, 3}
s2 = {1, 2, 4}
intersection_set = s1.intersection(s2)
print(intersection_set)  # Output: {1, 2}

# Use intersection_update to update a set with the intersection
s1.intersection_update(s2)
print(s1)  # Output: {1, 2}

# Check if two sets are disjoint
s1 = {1, 2}
s2 = {1, 2, 4}
s3 = {5}
print(s1.isdisjoint(s2))  # Output: False
print(s1.isdisjoint(s3))  # Output: True

# Check if a set is a subset or superset of another
s1 = {1, 2}
s2 = {1, 2, 4}
print(s1.issubset(s2))  # Output: True
print(s2.issuperset(s1))  # Output: True
print(s1.issuperset(s2))  # Output: False

# Find the symmetric difference of two sets
s1 = {1, 2}
s2 = {1, 2, 4}
symmetric_diff_set = s1.symmetric_difference(s2)
print(symmetric_diff_set)  # Output: {4}

# Find the union of two sets
union_set = s1.union(s2)
print(union_set)  # Output: {1, 2, 4}

# Update a set with the union of itself and another
s1.update(s2)
print(s1)  # Output: {1, 2, 4}
```
