# PYLINT

To use Pylint, you can install it using pip:

- pip install pylint

Once installed, you can run Pylint on your Python script from the command line:

- pylint your_script.py

or 

- pylint your_script.pu -ry (for detailed report)

### pylint test.py:

- This command runs pylint on the test.py file without any specific report options.
- It will generate a default set of reports, including a summary at the end of the analysis.
- The output may include messages about conventions, warnings, errors, and other information based on Pylint's analysis of the code.

### pylint test.py -ry:

- This command also runs pylint on the test.py file but includes the -r option with the y flag.
- The y flag specifies that raw messages should be displayed in the output.
- In this case, the output will provide a raw, machine-readable format of the messages generated by pylint. This format is often used for integration with other tools or for further processing.
